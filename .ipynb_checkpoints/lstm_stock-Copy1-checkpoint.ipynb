```python
ticker = yf.Ticker(stock)
splits = ticker.splits
splits.index = splits.index.tz_localize(None)

splits
```

```python
start_ts=pd.Timestamp(start).tz_localize(None)
end_ts=pd.Timestamp(end).tz_localize(None)

# Filter splits within range
splits_in_range=splits.loc[start_ts:end_ts]
```

```python
if not splits_in_range.empty:
    print(f"📉 Stock split history for {stock}:")
    print(splits)
else:
    print(f"No stock splits recorded for {stock}.")
```

```python
df['Date'] = pd.to_datetime(df['Date']).dt.tz_localize(None)
df_new = df[(df['Date'] >= start_ts) & (df['Date'] <= end_ts)]
```

```python
import matplotlib.dates as mdates
plt.figure(figsize=(15,9))
plt.plot(df_new['Date'], df['Close'],label='Closing Price',color='blue')
plt.title('Stock Closing Price Over Time')
plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # Tick each Year
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['SMA_10'] = df['Close'].rolling(window=100).mean()
df['SMA_20'] = df['Close'].rolling(window=50).mean()
```

```python
df['EMA_10'] = df['Close'].ewm(span=100, adjust=False).mean()
df['EMA_20'] = df['Close'].ewm(span=50, adjust=False).mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date'],df['Close'], label = f'{stock} Close Price', linewidth = 1)
plt.plot(df['Date'],df['SMA_10'] ,label = f'{stock} Moving Average 100 Price', linewidth = 1)
plt.plot(df['Date'],df['SMA_20'], label = f'{stock} Moving Average 50 Price', linewidth = 1)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date'],df['Close'], label = f'{stock} Close Price', linewidth = 1)
plt.plot(df['Date'],df['EMA_10'] ,label = f'{stock} Exponential Moving Average 100 Price', linewidth = 1)
plt.plot(df['Date'],df['EMA_20'], label = f'{stock} Exponential Moving Average 50 Price', linewidth = 1)
plt.legend()
plt.show()
```

```python
data_training = pd.DataFrame(df['Close'][:int(len(df)*0.75)])
data_testing = pd.DataFrame(df['Close'][int(len(df)*0.75):])
```

```python
data_training.shape
```

```python
data_testing.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array = scaler.fit_transform(data_training)
```

```python
data_training_array
```

```python
data_training_array.shape[0]
```

```python
x_train = []
y_train = []

for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100:i])
    y_train.append(data_training_array[i, 0])

x_train, y_train  = np.array(x_train), np.array(y_train)
```

```python
x_train.shape
```

```python
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
model = Sequential()

model.add(LSTM(units = 50, activation = 'relu', return_sequences = True, input_shape = (x_train.shape[1],1)))
model.add(Dropout(0.2))

model.add(LSTM(units = 60, activation = 'relu', return_sequences = True))
model.add(Dropout(0.3))

model.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
model.add(Dropout(0.4))

model.add(LSTM(units = 120, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(units = 1))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
model.fit(x_train, y_train, epochs = 50)
```

```python
past_100_days=data_training.tail(100)
```

```python
final_df = pd.concat([past_100_days, data_testing], ignore_index=True)
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
input_data = scaler.fit_transform(final_df)
```

```python
x_test = []
y_test = []

for i in range(100, input_data.shape[0]):
    x_test.append(input_data[i-100:i])
    y_test.append(input_data[i, 0])

x_test, y_test  = np.array(x_test), np.array(y_test)
```

```python
x_test.shape
```

```python
y_predicted = model.predict(x_test)
```

```python
y_predicted.shape
```

```python
scaler.scale_
```

```python
scaler_factor = 1 /scaler.scale_
y_predicted = y_predicted * scaler_factor
y_test = y_test * scaler_factor
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test, label = 'Original Price', linewidth = 1)
plt.plot(y_predicted, label = 'Predicted Price', linewidth = 1)
plt.legend()
plt.show()
```

```python
model.save('lstm_stock.h5')
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2 = r2_score(y_test, y_predicted)
mae = mean_absolute_error(y_test, y_predicted)
mse = mean_squared_error(y_test, y_predicted)

print(f"R2 Score: {r2:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
```
